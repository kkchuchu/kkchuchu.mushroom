```python
import numpy as np
import matplotlib.pyplot as plt
```

### Graph

```python
x_range = [-500., 1000.]
y_range = [-1.5e+11, 1.0e+11]
x_2 = np.arange(x_range[0], x_range[1], 0.2)
y_2 = -1 * pow(x_2, 4) + 1000 * pow(x_2, 3) - 20 * pow(x_2, 2) + 4 * x_2 - 6
plt.axis(x_range + y_range)
plt.plot(x_2, y_2)
```

```python
x_1 = np.arange(1, 100, 1)
y_1 = pow((x_1 %6), 2) % 7 - np.sin(x_1)
plt.plot(x_1, y_1)
```

### Optimize

```python
max(enumerate(y_1), key=lambda x: x[1])
```
